```python
from tkinter import Tk, Label, Entry, Button, messagebox

def on_validate_input(P):
    if P.replace('.', '', 1).isdigit() or P == "":
        return True
    else:
        messagebox.showerror("Invalid Input", "Please enter only digits.")
        return False

def validate_sum():
    inputs = [float(entry.get()) for entry in entry_boxes]
    total_sum = sum(inputs)

    if total_sum != 100:
        messagebox.showerror("Invalid Input", f"The sum of all integers must be 100 (current sum: {total_sum}).")
    else:
        global uw
        uw = inputs
        print(uw)
        messagebox.showinfo("Success", "Thank you for filling the form!")
        root.destroy()

# Create the main application window
root = Tk()
root.title("Assign Weightage to Critieria")


# Create a list to store the entry boxes and labels
entry_boxes = []
labels = ['Capital Cost', 'Operation and Maintaince Cost', 'Pollutant Removal', 'Micropollutant Removal', 'Eutrophication Potential', 'Carbon Sequestration', 'Land Area required to Build System', 'Simplicity' , 'Handle Variable Flow' , 'Nuisance to Local']

# Create input integer boxes with custom labels
for i in range(1, 11):
    label = Label(root, text=labels[i - 1], fg="black")
    label.grid(row=i, column=0, padx=5, pady=5)

    entry = Entry(root, width=5, validate="key")
    entry.grid(row=i, column=1, padx=5, pady=5)
    entry.config(validatecommand=(entry.register(on_validate_input), '%P'))
    entry_boxes.append(entry)

# Add the note label
note_label = Label(root, text="Criteria can be assigned equal weights, but the total of the numbers cannot exceed 100 (0 = unimportant, 100 = most important).", fg="black")
note_label.grid(row=0, column=0, columnspan=2, padx=5, pady=5)


# Create a submit button
submit_button = Button(root, text="Submit", command=validate_sum, fg="black")
submit_button.grid(row=11, column=0, columnspan=2, pady=10)

# Start the Tkinter event loop
root.mainloop()
```

```python
import tkinter as tk
import plotly.graph_objects as go

root = tk.Tk()

root.title("Appropriate Greywater Treatment Alternative")  # Adding the window title


categories = filteredlist
values = formattedfloatlist

width_size = 0

if len(categories) <= 3:
    width_size = 550
else:
    width_size = 1100

# Create a Plotly figure object
fig = go.Figure(data=[go.Bar(x=categories, y=values, marker_color='#1f77b4')]) 

fig.update_traces(marker_line_width=0.5, 
                  text=values,
                  textposition='outside',  
                  insidetextanchor='start',
                  hoverinfo='none')
    
# Update the layout of the figure object
fig.update_layout(title='Greywater Treatment Alternative Analysis',
                  xaxis_title='Technologies',
                  yaxis_title='Score',
                  width=width_size,
                  yaxis=dict(rangemode='nonnegative'),
                  xaxis_tickangle=0,
                  )


# Create a Tkinter canvas widget
canvas = tk.Canvas(root, width=1100, height=600)

# Create an image of the Plotly figure object
image = tk.PhotoImage(data=fig.to_image())

canvas.create_image(550, 300, image=image)

# Pack the canvas widget into the root window
canvas.pack()

# Start the Tkinter mainloop
root.mainloop()
```
